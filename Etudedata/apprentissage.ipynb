```python
#definition des series de valeurs des abscisses et des ordonness
x = df["brand_name"].values
y = df["price"].values
plt.plot(x,y)
plt.title("courbe des prix des smartphones depuis 2013")
plt.xlabel("marque")# icic c est la titre des abscisses
plt.ylabel("prix")#ici le titre des ordonees
plt.xticks(rotation=45)  # pour incliner les dates
plt.tight_layout()
plt.show()
# Utilisation de matplolib pour creer un graphique de type ligne
```

```python
#affichage des prix moyens des smartphones en fonction des mmarques grace au graphique de type bar
# on calcul donc les moyennes des prix de chaque marque et nom de tout l ensemble des marques confondus et sort_values sert a trier de manieres ascendant
prix_moyen = df.groupby("brand_name")["price"].mean().sort_values(ascending=True)
plt.bar(prix_moyen.index,prix_moyen.values)
plt.title('nnnnnnn')
plt.xlabel('marques')
plt.ylabel('prix moyen par marque')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#ressortir la relation entre deux variables grace au graphique de type scatter
# on calcul donc les moyennes des prix de chaque marque et nom de tout l ensemble des marques confondus et sort_values sert a trier de manieres ascendant
x = df["brand_name"].values
y = df["price"].values
plt.plot(x,y)
plt.scatter(x,y)
plt.title('nnnnnnn')
plt.xlabel('marques')
plt.ylabel('prix moyen par marque')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
